```python
import pandas as pd

# Load the data from a .tsv file into a DataFrame
df = pd.read_csv("data-2.tsv", delimiter='\t')
df = df.join(df.pop('primaryProfession').str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('primaryProfession'))
df = df.join(df.pop('knownForTitles').str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('knownForTitles'))

# Write the DataFrame to a new .csv file
#df.to_csv("data-2.csv", index=False)
```

```python
df = df.fillna(value=0)

# Remove duplicate rows
df = df.drop_duplicates()

# Write the DataFrame to a new .csv file
df.to_csv("data-2-clean.csv", index=False)
```

```python
# Load data into a pandas DataFrame
df = pd.read_csv('data-2-clean.csv')

# Group the data by nconst and apply the join aggregation to string columns
grouped = df.groupby('nconst').agg({'primaryName': 'first',
                                     'birthYear': 'first',
                                     'deathYear': 'first',
                                     'primaryProfession': ', '.join,
                                     'knownForTitles': ', '.join})

# Reset the index
grouped.reset_index(inplace=True)

# Save the resulting DataFrame to a new csv file
grouped.to_csv('grouped-data.csv', index=False)
```
