```python
import pandas as pd

# Load the data into a pandas DataFrame
df = pd.read_table("data-3.tsv", low_memory=False)
df=df.infer_objects()
```

```python
df.head()
```

```python
# Remove any duplicate rows from the DataFrame
df = df.drop_duplicates()

# Replace any missing values in the "ordering" column with 0
df["ordering"].fillna(0, inplace=True)

# Replace any missing values in the "language" column with "unknown"
df["language"].fillna("unknown", inplace=True)

# Remove any rows where the "title" column is missing
df = df[df["title"].notna()]

# Remove any rows where the "isOriginalTitle" column is missing
df = df[df["isOriginalTitle"].notna()]

# Convert the "isOriginalTitle" column to a Boolean data type
df["isOriginalTitle"] = df["isOriginalTitle"] == 1

# Select only the columns that match the table structure
df = df[["titleId", "ordering", "title", "region", "language", "types", "attributes", "isOriginalTitle"]]
```

```python
df.head()
```

```python
# Export the cleaned data to a new CSV file
df.to_csv("cleaned-data-3.csv", index=False)
```
